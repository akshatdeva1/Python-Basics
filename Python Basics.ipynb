```python
#14. Write a program to reverse a string using a while loop.

s= input("Enter a string:")
reverse = ''
i=len(s)-1
while i>=0:
  reverse+=s[i]
  i-=1
print("Reversed string", reverse)
```

```python
#15. Write a program to calculate the factoroial of a number provided by the user using a while loop.

num=int(input("Enter a number:"))
factorial=1
i=1
while i<=num:
    factorial*=i
    i+=1
print("Factorial of the number",num, "is", factorial)
```

**Theoretical Question and Answers**

1. What is Python, and why is it popular?
  - Python is a high-level, interpreted programming language known for its simplicity, readability, and flexibility. It is popular due to:
    - easy to understand
    - strong community support
    - extensive standard libraries

2. What is an interpreter in Python?
  - An interpreter is a program that reads and executes code line by line, converting high level Python code into machine understandable instructions during runtime.

3. What are pre defined keywords in Python?
  - Pre defined keywords in Python are reserved words that have special meanings and are part of the language syntax, such as 'if', 'else', 'while', 'for', 'def' etc. They cannot be used as identifiers or variable names.

4. Can keywords be used as variable names
  - No, keywords cannot be used as variable names in Python. Attempting to do so results in a syntax error, as these words are reserved for the core language functionality.

5. What is mutability in Python
Mutability refers to whether an object's state can be changed after it is created. A mutable object can have its contents altered, while an immutable object's state cannot be changed once defined.

6. Why are lists mutable, but tuples are immutable
  - Lists are mutable, meaning elements can be added, removed, or changed in place. This is because lists are designed to support dynamic data manipulation.
  - Tuples are immutable, so once created, their elements cannot be altered, deleted, or added; this protects data integrity and allows tuples to be used as dictionary keys or set elements.

7. What is the difference between “==” and “is” operators in Python
  -   `==` (equality operator): Checks if the values of two objects are equal, regardless of whether they are the same object in memory.
  - `is` (identity operator): Checks if two variables reference the exact same object in memory, not just if their contents are equal.

8. What are logical operators in Python
  - Logical operators perform boolean logic operations:
  `and`: Returns True if both conditions are True.
  `or`: Returns True if at least one condition is True.
  `not`: Inverts the Boolean value.

9. What is type casting in Python
  - Type casting refers to converting one data type to another, either explicitly or implicitly.

10. What is the difference between implicit and explicit type casting?
  - Implicit type casting: The interpreter automatically converts one data type to another where appropriate.
  - Explicit type casting: The programmer manually converts the data type using functions such as `int()`, `float()`, or `str()`.

11. What is the purpose of conditional statements in Python
  - Conditional statements control program flow by executing different blocks of code based on whether a condition is True or False (e.g., using `if`, `elif`, and `else`).

12. How does the elif statement work?
  - `elif` (“else if”) allows multiple conditions to be checked sequentially. It runs its code block only if preceding `if` and `elif` conditions are False and its own condition is True.

13. What is the difference between for and while loops?
  - for loop: Iterates over a sequence (like a list, tuple, or range) a specific number of times.
  -	while loop: Continues iterating as long as a given condition remains True, regardless of the number of iterations.

14. Describe a scenario where a while loop is more suitable than a for loop
  - A while loop is better suited when the number of iterations is not known in advance but depends on a dynamically changing condition—such as reading input until the user enters a specific value or waiting for a process to complete.
